```python
# starting model creation and training
#using linear regression
#to use linear regression we need finite distinct value (for that we need to convert date formate to days )
data['days']=(data.index - data.index.min()).days  # finding the minimum date and minusing it from the other dates to find the days (like 1day,2day, 3day ....)
#regression is only applicable for numerical values .days convertes it to numerical values
```

```python
x=data['days'].values.reshape(-1,1)#  .values function returns a numpy array and .reshape makes it as a 2d array ,  (-1,1)>> the -1 tells the function to find the appropriate row size while keeping 1 coloumn
y=data['Close'].values.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(data['days'].isnull().sum())
print(data['Close'].isnull().sum())
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
#predicting
y_pred=model.predict(x_test)

#ploting actual vs predicted value
plt.figure(figsize=(10,10))
plt.scatter(x_test,y_test,color='green',label='Actual value')
plt.plot(x_test,y_pred,color='blue',linewidth=2,label='Predicted value ')
plt.xlabel('days')
plt.ylabel('closing price')
plt.title('Stock Price Prediction using Linear Regression',color='red')
plt.legend(loc='upper left')
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
r2=r2_score(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
mea=mean_absolute_error(y_test,y_pred)
print(mea)
print(mse)
rmse = np.sqrt(mse)
print("RMSE:", rmse)
print('r2 score',r2)
```
