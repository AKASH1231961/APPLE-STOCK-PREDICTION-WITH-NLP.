```python
# starting model creation and training
#using linear regression
#to use linear regression we need finite distinct value (for that we need to convert date formate to days )
data['days']=(data.index - data.index.min()).days  # finding the minimum date and minusing it from the other dates to find the days (like 1day,2day, 3day ....)
#regression is only applicable for numerical values .days convertes it to numerical values
```

```python
x=data['days'].values.reshape(-1,1)#  .values function returns a numpy array and .reshape makes it as a 2d array ,  (-1,1)>> the -1 tells the function to find the appropriate row size while keeping 1 coloumn
y=data['Close'].values.reshape(-1,1)
```
